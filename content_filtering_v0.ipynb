```python
import numpy as np
import pandas as pd
```

```python
columns = ['genres', 'id',]
movie_df = pd.read_csv("./processed/movies_processed.csv", usecols=columns).sample(n=100)
```

```python
genres_df = pd.read_csv('./processed/genres.csv')
genres_df
```

```python
from ast import literal_eval
```

```python
def genre_mask(genres):
    genre_mask = [0]*len(genres_df.index)
    genres = literal_eval(genres)
    for genre in genres:
        idx = genres_df.loc[genres_df['genre_id']==genre].index.values[0]
        genre_mask[idx] = 1
    return genre_mask
```

```python
new_df_cols = ['id']+genres_df.genre_name.values.tolist()
# print(new_df_cols)
new_df = pd.DataFrame(columns=new_df_cols)
for index, row in movie_df.iterrows():
    # print(type(row))
    # print(row)
    genres = genre_mask(row['genres'])
    genres = [row.id]+genres
    # print( genres )
    new_df = new_df.append(pd.DataFrame( [genres],
               columns=new_df_cols,),
               ignore_index = True,)
    # pd.DataFrame( [genres], columns=new_df_cols,)
    # new_df.iloc[index] = genres
    # break
```
